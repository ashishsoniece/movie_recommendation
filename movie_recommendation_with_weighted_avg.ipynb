```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import os
print(os.listdir())
```

```python
credits = pd.read_csv("tmdb_5000_credits.csv")
movies = pd.read_csv("tmdb_5000_movies.csv")
```

```python
credits.head()
```

```python
movies.head()
```

```python
movie_id_rename = credits.rename(index = str,
                                 columns = {"movie_id":"id"})
movie_merged = movies.merge(movie_id_rename, on= "id")
movie_merged.head()
```

```python
movie_merged.shape
```

```python
movie_merged.info()
```

```python
movie_merged.columns
```

```python
movie_cleaned = movie_merged.drop(columns= ["homepage",'title_x','title_y','production_countries','status'],axis = 1)
movie_cleaned.head()
```

```python
R = movie_cleaned['vote_average']
v = movie_cleaned['vote_count']
C = movie_cleaned['vote_average'].mean()
m = movie_cleaned['vote_count'].quantile(0.70)
```

```python
movie_cleaned['weighted_avg'] = ((R*v)+(C*m))/(v+m)
```

```python
movie_cleaned.head()
```

```python
weighted_sorted_avg = movie_cleaned.sort_values(by= 'weighted_avg',
                                                ascending= False)
weighted_sorted_avg[['original_title','vote_count','vote_average','weighted_avg','popularity']].head(10)
```

```python
axis1 = sns.barplot(x = weighted_sorted_avg['weighted_avg'].head(10),
                    y = weighted_sorted_avg['original_title'].head(10))
plt.xlim(4,10)
plt.xlabel('weighted_avg', weight = 'bold')
plt.ylabel('Original Title',weight = 'bold')
plt.title('Best movies by vote average',weight = 'bold')
plt.show()
```

```python
popularity = weighted_sorted_avg.sort_values(by= ['popularity','weighted_avg'],
                                             ascending= False)
popularity[['original_title','vote_count','vote_average','weighted_avg','popularity']].head(10)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
movie_scaled = scaler.fit_transform(movie_cleaned[['weighted_avg','popularity']])
normalised_mov_df = pd.DataFrame(movie_scaled,columns=['weighted_avg','popularity'])
normalised_mov_df.head()
```

```python
movie_cleaned[['normalised_weighted_avg','normalised_popularity']] = normalised_mov_df
movie_cleaned.head()
```

```python
movie_cleaned['score'] = movie_cleaned['normalised_weighted_avg']*0.5 + movie_cleaned['normalised_popularity']*0.5
normalised_sort_movies = movie_cleaned.sort_values('score',ascending=False)
normalised_sort_movies[['original_title',
                        'vote_count',
                        'vote_average',
                        'normalised_weighted_avg',
                        'normalised_popularity',
                        'score']].head(10)
```
